```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis,zscore

#Step 1: Load the Kaggle dataset
data=pd.read_csv('D:/supermarket_sales - Sheet1.csv')  

#Display the first few rows to confirm it loaded correctly
# print(data.head())

#Step 2: Select the "Total gross income" coulumn
income=data['Total Gross Income']


#Step 3:Statistical Calculation
mean_val=income.mean()
median_val=income.median()
mode_val=income.mode().iloc[0] if not income.mode().empty else np.nan
std_dev=income.std()
variance=income.var()
kurt=kurtosis(income)
cv=std_dev/mean_val #Coefficient of Varitaion
z_scores=zscore(income)


#Step 4: Display the result
print("Statistical Summary of 'Total Gross Income:'")
print(f"Mean:{mean_val:.2f}")
print(f"Median:{median_val:.2f}")
print(f"Mode:{mode_val:.2f}")
print(f"Standard Deviation:{std_dev:.2f}")
print(f"Variance:{variance:.2f}")
print(f"Kurtosis:{kurt:.2f}")
print(f"Coefficient of Variaton:{cv:.2f}")
print("\n First 5 Z_scores:\n:",z_scores[:5])

#Histogram for distributions
plt.figure(figsize=(10,6))
sns.histplot(income, kde=True)
plt.title("Distribution of Total Gross Income")
plt.xlabel("Total Gross Income")
plt.ylabel("Frequency")
plt.show()

#BoxPlot to check outliers and spread
plt.figure(figsize=(8,6))
sns.boxplot(x=income)
plt.title("Boxplot of Total Gross Income")
plt.xlabel("Total Gross Income")
plt.show()
```
